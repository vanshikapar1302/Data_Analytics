```python
import numpy as np
import pandas  as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Employee.csv")
```

```python
df.head()
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
# Age ka mean nikal rahe hain
age_mean = df['Age'].mean()

# Missing Age ko mean se fill karke wapas column me assign
df['Age'] = df['Age'].fillna(age_mean)
```

```python
# Salary ka median nikal rahe hain (outliers ke liye better)
salary_median = df['Salary'].median()

# Missing salary ko median se fill
df['Salary']=df['Salary'].fillna(salary_median)
```

```python
# Categorical columns identify kar rahe hain
cat_cols = df.select_dtypes(include='object').columns
cat_cols
```

```python
from sklearn.preprocessing import LabelEncoder

# Har categorical column pe label encoding
le = LabelEncoder()

for col in cat_cols:
    df[col] = le.fit_transform(df[col])
```
